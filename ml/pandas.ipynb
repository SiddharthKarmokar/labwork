```python
import pandas as pd
n = 6
data = []
for _ in range(n):
    a, b = input().split()
    data.append((a, int(b)))
df = pd.DataFrame(data, columns=['department', 'salary'])
result = (
    df.groupby('department', as_index=False)['salary']
    .mean()
    .rename(columns={'salary':'avg_salary'})
    )
result
```
