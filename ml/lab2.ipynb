```python
import pandas as pd

def main()->None:
    n = int(input())
    names, math_score, science_score = [], [], []
    for _ in range(n):
        name, math, science = input().split()
        math, science = int(math), int(science)
        names.append(name)
        math_score.append(math)
        science_score.append(science)
    data = {
        "name": names,
        "math_score": math,
        "science_score": science
    }
    df = pd.DataFrame(data=data)
    df["total_score"] = df["math_score"] + df["science_score"]
    print(df)
    return
    
    
    

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import sys 

def main()->None:
    n = int(input())
    names, salaries = [], []
    for _ in range(n):
        name, salary = input().split()
        salary = int(salary)
        names.append(name)
        salaries.append(salary)
    threshold = int(input())
    data = {
        "name": names,
        "salary": salaries
    }
    df = pd.DataFrame(data=data)
    res = df[df["salary"] > threshold]
    print(res)
    return

if __name__ == "__main__":
    try:
        sys.stdin = open("input.txt")
    except:
        pass
    main()
```

```python
import pandas as pd
import sys 

def main()->None:
    n = int(input())
    data = []
    for _ in range(n):
        region, product, price = input().split()
        price = int(price)
        data.append((region, product, price))
    df = pd.DataFrame(data, columns=["region", "product", "price"])
    res = pd.pivot_table(
        df,
        values='price',
        index=['region'],
        columns=['product']
    ).fillna(0).astype(int)
    print(res)
    return

if __name__ == "__main__":
    try:
        sys.stdin = open("input.txt")
    except:
        pass
    main()
```

```python
import pandas as pd
import sys 

def main()->None:
    n = int(input())
    data = []
    for _ in range(n):
        oid, cid, amt = map(int, input().split())
        data.append((oid, cid, amt))
    df = pd.DataFrame(data, columns=['order_id', 'customer_id', 'amount'])
    df1 = df.groupby(["customer_id"], as_index=False)["amount"].sum()
    df1.rename(columns={"amount": "total_amount"}, inplace=True)
    m = int(input())
    data = []
    for _ in range(m):
        cid, name, city = input().split()
        cid = int(cid)
        data.append((cid, name, city))
    df2 = pd.DataFrame(data, columns=['customer_id', 'name', 'city'])

    merged = pd.merge(df1, df2, on="customer_id")
    res = merged[merged["total_amount"] > 500]
    neworder = ["customer_id", "name", "city", "total_amount"]
    res = res[neworder]
    res.reset_index(drop=True, inplace=True)
    res.sort_values(by='total_amount', ascending=False)
    print(res)
    return

if __name__ == "__main__":
    try:
        sys.stdin = open("input.txt")
    except:
        pass
    main()
```
