```python
import numpy as np

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def sigmoid_derivative(a):
    return a * (1.0 - a)

X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]], dtype=float)

Y = np.array([[0,0,1,0],
              [0,1,1,1],
              [0,1,1,1],
              [1,1,0,0]], dtype=float)

n_in, n_hidden, n_out = 2, 3, 4
np.random.seed(42)

W1 = np.random.uniform(-1, 1, (n_in, n_hidden)) * 0.5
b1 = np.zeros((1, n_hidden))
W2 = np.random.uniform(-1, 1, (n_hidden, n_out)) * 0.5
b2 = np.zeros((1, n_out))

lr = 0.3
max_epochs = 10000
tol = 1e-2
m = X.shape[0]
MSE_div = m * n_out

for epoch in range(1, max_epochs + 1):
    z1 = X @ W1 + b1
    a1 = sigmoid(z1)
    z2 = a1 @ W2 + b2
    yhat = sigmoid(z2)

    loss = np.sum((Y - yhat) ** 2) / MSE_div

    dL_dyhat = (yhat - Y) * (2.0 / MSE_div)
    delta2 = dL_dyhat * sigmoid_derivative(yhat)
    dW2 = a1.T @ delta2
    db2 = np.sum(delta2, axis=0, keepdims=True)
    delta1 = (delta2 @ W2.T) * sigmoid_derivative(a1)
    dW1 = X.T @ delta1
    db1 = np.sum(delta1, axis=0, keepdims=True)

    W2 -= lr * dW2
    b2 -= lr * db2
    W1 -= lr * dW1
    b1 -= lr * db1

    if epoch % 1000 == 0:
        print(f"Epoch: {epoch}, Loss: {loss:.4f}")
    if loss < tol:
        print(f"Converged at epoch {epoch}, Loss: {loss:.4f}")
        break

yhat_final = sigmoid(sigmoid(X @ W1 + b1) @ W2 + b2)
print("\nPredictions (raw and rounded):")
for x, y_true, y_pred in zip(X, Y, yhat_final):
    print(f"{x.tolist()} -> true {y_true.tolist()} ; "
          f"pred {[round(v) for v in y_pred.tolist()]} "
          f"(raw {y_pred.round(3).tolist()})")
```
