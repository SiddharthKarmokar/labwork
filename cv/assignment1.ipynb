### Question 7

```python
gray = cv.cvtColor(X, cv.COLOR_RGB2GRAY)

gray1d = gray.ravel()
plt.hist(gray1d, bins=256, range=[0, 256], color='gray')
plt.xlabel("Pixel values")
plt.ylabel("Frequency")
plt.show()
```

### Question 8

```python
f32 = np.array(X).astype(np.float32)
meanval = np.mean(f32)
stdval = np.std(f32)
print(meanval, stdval)
```

```python
norm = (f32 - meanval)/stdval
plt.imshow(norm)
```

### Question 9

```python
nimg = X.copy()
dim = X.shape
for i in range(dim[0]):
    for j in range(dim[1]):
        nimg[i, j] = 255 - nimg[i, j]


plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.imshow(nimg)
plt.subplot(1, 2, 2)
plt.imshow(X)
```

### Question 10

```python
nimg = X.copy()
dim = X.shape
for i in range(dim[0]):
    for j in range(dim[1]):
        if np.any(nimg[i, j] > 127):
            nimg[i, j] = [255, 255, 255]
        else:
            nimg[i, j] = [0, 0, 0]

plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.imshow(nimg)
plt.subplot(1, 2, 2)
plt.imshow(X)
```
